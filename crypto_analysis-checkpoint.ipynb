```python
# Moving Average Crossover Trading Strategy
def ma_crossover_signals(data, short_window=20, long_window=50):
    """
    Generate buy and sell signals based on moving average crossover
    
    Parameters:
    data (pd.Series): Price data
    short_window (int): Short-term moving average window
    long_window (int): Long-term moving average window
    
    Returns:
    pd.DataFrame: DataFrame with price data and signals
    """
    # Create a DataFrame with price data
    signals = pd.DataFrame(index=data.index)
    signals['price'] = data
    
    # Create short and long moving averages
    signals['short_ma'] = data.rolling(window=short_window).mean()  # Calculate short-term moving average
    signals['long_ma'] = data.rolling(window=long_window).mean()    # Calculate long-term moving average
    
    # Create signals
    signals['signal'] = 0.0  # Initialize signal column with zeros
    # Generate buy signal (1) when short MA crosses above long MA
    # We start from short_window index since earlier data points don't have enough history for the short MA
    signals['signal'][short_window:] = np.where(
        signals['short_ma'][short_window:] > signals['long_ma'][short_window:], 1.0, 0.0)
    
    # Generate trading orders
    # diff() calculates the difference between current and previous signal
    # This creates:
    # 1.0 when position changes from 0 to 1 (buy signal)
    # -1.0 when position changes from 1 to 0 (sell signal)
    # 0.0 when position doesn't change
    signals['positions'] = signals['signal'].diff()
    
    # Calculate strategy returns
    # Multiply price changes by positions to get returns from the strategy
    signals['price_change'] = signals['price'].pct_change()
    signals['strategy_returns'] = signals['price_change'] * signals['signal'].shift(1)
    
    # Calculate trade statistics
    # Identify individual trades (from buy to sell)
    signals['trade_id'] = (signals['positions'] != 0).cumsum()
    
    # Calculate trade-specific metrics
    # This can be used to analyze individual trade performance
    signals['in_trade'] = signals['signal']
    
    return signals

# Example usage:
# signals = ma_crossover_signals(closing_prices, short_window=20, long_window=50)
# buy_signals = signals[signals['positions'] == 1.0]
# sell_signals = signals[signals['positions'] == -1.0]

# Further analysis examples:
# 1. Count total trades: len(buy_signals)
# 2. Calculate cumulative returns: (1 + signals['strategy_returns']).cumprod()
# 3. Find max profit trade: signals.groupby('trade_id')['strategy_returns'].sum().max()
# 4. Find max loss trade: signals.groupby('trade_id')['strategy_returns'].sum().min()
# 5. Calculate win rate: (signals.groupby('trade_id')['strategy_returns'].sum() > 0).mean()
# 6. Calculate average profit per trade: signals.groupby('trade_id')['strategy_returns'].sum().mean()
```

```python
# RSI (Relative Strength Index) calculation on Simple Moving Average
def calculate_rsi_on_sma(data, sma_window=14, rsi_window=14):
    """
    Calculate RSI (Relative Strength Index) on Simple Moving Average
    
    Parameters:
    data (pd.Series): Price data
    sma_window (int): Window for Simple Moving Average calculation
    rsi_window (int): Window for RSI calculation
    
    Returns:
    pd.DataFrame: DataFrame with price data, SMA and RSI values
    """
    # Create a DataFrame for results
    results = pd.DataFrame(index=data.index)
    results['price'] = data
    
    # Calculate Simple Moving Average
    results['sma'] = data.rolling(window=sma_window).mean()
    
    # Calculate price changes of the SMA
    results['sma_change'] = results['sma'].diff()
    
    # Separate gains and losses
    results['gain'] = np.where(results['sma_change'] > 0, results['sma_change'], 0)
    results['loss'] = np.where(results['sma_change'] < 0, -results['sma_change'], 0)
    
    # Calculate average gains and losses over the RSI window
    results['avg_gain'] = results['gain'].rolling(window=rsi_window).mean()
    results['avg_loss'] = results['loss'].rolling(window=rsi_window).mean()
    
    # Alternative method using exponential moving average for smoother RSI
    # results['avg_gain'] = results['gain'].ewm(span=rsi_window, adjust=False).mean()
    # results['avg_loss'] = results['loss'].ewm(span=rsi_window, adjust=False).mean()
    
    # Calculate RS (Relative Strength)
    # Add small epsilon to avoid division by zero
    epsilon = 1e-10
    results['rs'] = results['avg_gain'] / (results['avg_loss'] + epsilon)
    
    # Calculate RSI
    results['rsi'] = 100 - (100 / (1 + results['rs']))
    
    # Clean up intermediate columns if needed
    # results = results[['price', 'sma', 'rsi']]
    
    return results

# Example usage:
# df = calculate_rsi_on_sma(closing_prices, sma_window=20, rsi_window=14)
# Overbought condition: df[df['rsi'] > 70]
# Oversold condition: df[df['rsi'] < 30]
```

```python
import pandas as pd
import numpy as np

def calculate_rsi_on_sma(data, sma_window, rsi_window):
    """
    Calculate RSI (Relative Strength Index) on Simple Moving Average.
    
    This function first calculates a Simple Moving Average (SMA) on the price data,
    then computes the RSI indicator based on the changes in the SMA values.
    
    Parameters:
        data (pd.Series): Price data time series
        sma_window (int): Window size for Simple Moving Average calculation
        rsi_window (int): Window size for RSI calculation
    
    Returns:
        pd.DataFrame: DataFrame containing:
            - price: Original price data
            - sma: Simple Moving Average values
            - sma_change: Change in SMA values
            - gain: Positive changes in SMA
            - loss: Absolute value of negative changes in SMA
            - avg_gain: Average of gains over RSI window
            - avg_loss: Average of losses over RSI window
            - rs: Relative Strength (avg_gain/avg_loss)
            - rsi: Relative Strength Index values (0-100)
    """
    # Create a DataFrame for results
    results = pd.DataFrame(index=data.index)
    results['price'] = data
    
    # Calculate Simple Moving Average
    results['sma'] = data.rolling(window=sma_window).mean()
    
    # Calculate price changes of the SMA
    results['sma_change'] = results['sma'].diff()
    
    # Separate gains and losses
    results['gain'] = np.where(results['sma_change'] > 0, results['sma_change'], 0)
    results['loss'] = np.where(results['sma_change'] < 0, -results['sma_change'], 0)
    
    # Calculate average gains and losses over the RSI window
    results['avg_gain'] = results['gain'].rolling(window=rsi_window).mean()
    results['avg_loss'] = results['loss'].rolling(window=rsi_window).mean()
    
    # Calculate RS (Relative Strength) with epsilon to avoid division by zero
    epsilon = 1e-10
    results['rs'] = results['avg_gain'] / (results['avg_loss'] + epsilon)
    
    # Calculate RSI
    results['rsi'] = 100 - (100 / (1 + results['rs']))
    
    return results
```

```python
import numpy as np
import pandas as pd
from typing import Union, List, Optional


def simple_moving_average(data: Union[List[float], np.ndarray, pd.Series], 
                          window_size: int) -> Union[np.ndarray, pd.Series]:
    """
    Calculate the Simple Moving Average (SMA) for a given dataset.
    
    Args:
        data: Time series data as a list, numpy array, or pandas Series
        window_size: Number of periods to include in the moving average
        
    Returns:
        Array or Series containing the moving averages with NaN values for the first (window_size-1) elements
    """
    if isinstance(data, pd.Series):
        return data.rolling(window=window_size).mean()
    else:
        data_array = np.array(data)
        result = np.full_like(data_array, np.nan, dtype=float)
        for i in range(window_size - 1, len(data_array)):
            result[i] = np.mean(data_array[i - window_size + 1:i + 1])
        return result


def exponential_moving_average(data: Union[List[float], np.ndarray, pd.Series], 
                               span: int, 
                               adjust: bool = True) -> Union[np.ndarray, pd.Series]:
    """
    Calculate the Exponential Moving Average (EMA) for a given dataset.
    
    Args:
        data: Time series data as a list, numpy array, or pandas Series
        span: Number of periods to use for calculating the smoothing factor
        adjust: If True, use the traditional adjustment to startup periods
        
    Returns:
        Array or Series containing the exponential moving averages
    """
    if isinstance(data, pd.Series):
        return data.ewm(span=span, adjust=adjust).mean()
    else:
        # Convert to pandas Series for calculation then back to numpy if needed
        series_data = pd.Series(data)
        result = series_data.ewm(span=span, adjust=adjust).mean()
        return result.values if not isinstance(data, pd.Series) else result


def weighted_moving_average(data: Union[List[float], np.ndarray, pd.Series], 
                            window_size: int,
                            weights: Optional[Union[List[float], np.ndarray]] = None) -> Union[np.ndarray, pd.Series]:
    """
    Calculate the Weighted Moving Average (WMA) for a given dataset.
    
    Args:
        data: Time series data as a list, numpy array, or pandas Series
        window_size: Number of periods to include in the moving average
        weights: Custom weights to apply. If None, linear weights are used
                (more recent values get higher weights)
        
    Returns:
        Array or Series containing the weighted moving averages with NaN values for the first (window_size-1) elements
    """
    if weights is None:
        # Default to linear weights (1, 2, 3, ..., window_size)
        weights = np.arange(1, window_size + 1)
    
    # Normalize weights to sum to 1
    weights = np.array(weights) / np.sum(weights)
    
    if isinstance(data, pd.Series):
        return data.rolling(window=window_size).apply(
            lambda x: np.sum(weights * x) if len(x) == window_size else np.nan, raw=True)
    else:
        data_array = np.array(data)
        result = np.full_like(data_array, np.nan, dtype=float)
        for i in range(window_size - 1, len(data_array)):
            window = data_array[i - window_size + 1:i + 1]
            result[i] = np.sum(weights * window)
        return result


def cumulative_moving_average(data: Union[List[float], np.ndarray, pd.Series]) -> Union[np.ndarray, pd.Series]:
    """
    Calculate the Cumulative Moving Average (CMA) for a given dataset.
    
    Args:
        data: Time series data as a list, numpy array, or pandas Series
        
    Returns:
        Array or Series containing the cumulative moving averages
    """
    if isinstance(data, pd.Series):
        return data.expanding().mean()
    else:
        data_array = np.array(data)
        result = np.zeros_like(data_array, dtype=float)
        cumsum = 0
        for i in range(len(data_array)):
            cumsum += data_array[i]
            result[i] = cumsum / (i + 1)
        return result


# Example usage
if __name__ == "__main__":
    # Sample data
    data = [10, 15, 20, 25, 30, 35, 40, 45, 50]
    
    # Calculate different types of moving averages
    sma = simple_moving_average(data, window_size=3)
    ema = exponential_moving_average(data, span=3)
    wma = weighted_moving_average(data, window_size=3)
    cma = cumulative_moving_average(data)
    
    # Create a DataFrame to display results
    results = pd.DataFrame({
        'Data': data,
        'SMA(3)': sma,
        'EMA(3)': ema,
        'WMA(3)': wma,
        'CMA': cma
    })
    
    print(results)
```

```python
def install_packages_from_list(packages_list, requirements_file=False, upgrade=False, pip_options=None):
    """
    Install Python packages using pip with improved error handling
    
    Args:
        packages_list: String or list of package names, or path to requirements file
        requirements_file: Boolean indicating if packages_list is a requirements file
        upgrade: Boolean indicating if packages should be upgraded
        pip_options: Dictionary of additional pip options as key-value pairs
    
    Returns:
        None
    
    Raises:
        FileNotFoundError: If requirements_file is True and file doesn't exist
        subprocess.CalledProcessError: If pip installation fails
    """
    # Build the base pip command
    pip_command = [sys.executable, "-m", "pip", "install"]
    
    # Add upgrade flag if specified
    if upgrade:
        pip_command.append("--upgrade")
    
    # Add any additional pip options
    if pip_options:
        for option, value in pip_options.items():
            pip_command.extend([f"--{option}", value])
    
    # Handle requirements file
    if requirements_file:
        if isinstance(packages_list, str):
            if not os.path.exists(packages_list):
                raise FileNotFoundError(f"Requirements file not found: {packages_list}")
            pip_command.extend(["-r", packages_list])
            print(f"Installing packages from requirements file: {packages_list}")
            try:
                # Added error handling with verbose output
                result = subprocess.run(pip_command, check=False, capture_output=True, text=True)
                if result.returncode != 0:
                    print(f"Error installing from requirements file: {result.stderr}")
                    print("Continuing with other installations...")
                else:
                    print("Requirements file installation successful")
            except Exception as e:
                print(f"Exception during requirements installation: {str(e)}")
        else:
            raise TypeError("When requirements_file=True, packages_list must be a string path")
    
    # Handle list of packages
    else:
        if isinstance(packages_list, str):
            packages = [packages_list]
        else:
            packages = packages_list
            
        for package in packages:
            print(f"Installing package: {package}")
            package_command = pip_command + [package]
            try:
                # Added error handling with verbose output
                result = subprocess.run(package_command, check=False, capture_output=True, text=True)
                if result.returncode != 0:
                    print(f"Error installing {package}: {result.stderr}")
                    print(f"You may need to install {package} manually or check for conflicts")
                else:
                    print(f"Successfully installed {package}")
            except Exception as e:
                print(f"Exception during {package} installation: {str(e)}")
    
    print("Package installation process completed!")


# Example usage:
if __name__ == "__main__":
    # Install individual packages with better error handling
    install_packages_from_list(["numpy", "pandas", "matplotlib"], upgrade=True)
    
    # Install from requirements.txt
    try:
        install_packages_from_list("requirements.txt", requirements_file=True)
    except FileNotFoundError:
        print("Requirements file not found, skipping...")
    
    # Install with additional pip options
    install_packages_from_list(
        ["tensorflow"], 
        pip_options={"index-url": "https://pypi.org/simple/"}
    )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Create or load your data
# -----------------------------
dates = pd.date_range(start='2022-01-01', periods=300)
np.random.seed(42)

# Simulated OHLC data
high = np.random.normal(0, 1, 300).cumsum() + 100
low = high - np.random.uniform(1, 5, 300)
close = low + np.random.uniform(0, 5, 300)

df = pd.DataFrame({
    'High': high,
    'Low': low,
    'Close': close
}, index=dates)

# -----------------------------
# 2. Calculate CCI manually
# -----------------------------
def calculate_cci(df, period=20):
    df = df.copy()
    
    # Typical Price
    df['TP'] = (df['High'] + df['Low'] + df['Close']) / 3
    
    # SMA of TP
    df['SMA_TP'] = df['TP'].rolling(period).mean()
    
    # Mean Deviation
    df['MeanDev'] = df['TP'].rolling(period).apply(
        lambda x: np.mean(np.abs(x - np.mean(x)))
    )
    
    # CCI formula
    df['CCI'] = (df['TP'] - df['SMA_TP']) / (0.015 * df['MeanDev'])
    
    return df

df = calculate_cci(df)

# -----------------------------
# 3. Detect Divergence
# -----------------------------
def detect_divergence(df, price_col='Close', indicator_col='CCI', window=5):
    df = df.copy()
    
    df['price_high'] = False
    df['price_low'] = False
    df['indicator_high'] = False
    df['indicator_low'] = False
    
    # Local highs/lows
    for i in range(window, len(df) - window):
        price_slice = df[price_col].iloc[i-window:i+window+1]
        ind_slice = df[indicator_col].iloc[i-window:i+window+1]
        
        if df[price_col].iloc[i] == price_slice.max():
            df.loc[df.index[i], 'price_high'] = True
        if df[price_col].iloc[i] == price_slice.min():
            df.loc[df.index[i], 'price_low'] = True
        
        if df[indicator_col].iloc[i] == ind_slice.max():
            df.loc[df.index[i], 'indicator_high'] = True
        if df[indicator_col].iloc[i] == ind_slice.min():
            df.loc[df.index[i], 'indicator_low'] = True
    
    # Divergence detection
    df['bearish_div'] = False
    df['bullish_div'] = False
    
    for i in range(window*2, len(df)):
        # Bearish divergence: price HH, indicator LH
        if df['price_high'].iloc[i]:
            for j in range(i-1, max(0, i-window*4), -1):
                if df['price_high'].iloc[j]:
                    if (df[price_col].iloc[i] > df[price_col].iloc[j] and
                        df[indicator_col].iloc[i] < df[indicator_col].iloc[j]):
                        df.loc[df.index[i], 'bearish_div'] = True
                    break
        
        # Bullish divergence: price LL, indicator HL
        if df['price_low'].iloc[i]:
            for j in range(i-1, max(0, i-window*4), -1):
                if df['price_low'].iloc[j]:
                    if (df[price_col].iloc[i] < df[price_col].iloc[j] and
                        df[indicator_col].iloc[i] > df[indicator_col].iloc[j]):
                        df.loc[df.index[i], 'bullish_div'] = True
                    break
    
    return df

df = detect_divergence(df)

# -----------------------------
# 4. Plot Results
# -----------------------------
plt.figure(figsize=(14, 10))

# Price chart
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df.index, df['Close'], label='Close Price')

# Divergence markers
ax1.scatter(df.index[df['bearish_div']], df['Close'][df['bearish_div']],
            color='red', marker='v', s=120, label='Bearish Divergence')

ax1.scatter(df.index[df['bullish_div']], df['Close'][df['bullish_div']],
            color='green', marker='^', s=120, label='Bullish Divergence')

ax1.set_title("Price with CCI Divergence Signals")
ax1.grid(True)
ax1.legend()

# CCI chart
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
ax2.plot(df.index, df['CCI'], label='CCI', color='blue')
ax2.axhline(100, color='red', linestyle='--', alpha=0.4)
ax2.axhline(-100, color='green', linestyle='--', alpha=0.4)
ax2.axhline(0, color='black', linestyle='-', alpha=0.3)
ax2.set_title("Commodity Channel Index (CCI)")
ax2.grid(True)
ax2.legend()

# Add some improvements: highlight divergence points on CCI chart too
for idx in df.index[df['bearish_div']]:
    ax2.scatter(idx, df.loc[idx, 'CCI'], color='red', marker='v', s=120)
    
for idx in df.index[df['bullish_div']]:
    ax2.scatter(idx, df.loc[idx, 'CCI'], color='green', marker='^', s=120)

plt.tight_layout()
plt.show()

# Add summary statistics
print(f"Total bearish divergences detected: {df['bearish_div'].sum()}")
print(f"Total bullish divergences detected: {df['bullish_div'].sum()}")

# Show a sample of the divergence points
print("\nSample of divergence points:")
div_sample = df[df['bearish_div'] | df['bullish_div']].head(5)
print(div_sample[['Close', 'CCI', 'bearish_div', 'bullish_div']])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample price data
dates = pd.date_range(start='2022-01-01', periods=200)
close_prices = np.random.normal(0, 1, 200).cumsum() + 100
df = pd.DataFrame({'Close': close_prices}, index=dates)

# RSI Calculation Function
def compute_RSI(series, period=14):
    delta = series.diff()
    
    # Gains (positive deltas) and losses (negative deltas)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    # Average gain and loss (Wilder's smoothing)
    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()
    
    # Relative Strength (RS)
    RS = avg_gain / avg_loss
    
    # RSI formula
    RSI = 100 - (100 / (1 + RS))
    
    return RSI

# Apply RSI to DataFrame
df['RSI'] = compute_RSI(df['Close'], period=14)

# Create plot
plt.figure(figsize=(12, 8))

# Price chart
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df.index, df['Close'], label='Close Price')
ax1.set_title("Price and 14Period RSI")
ax1.grid(True)
ax1.legend()

# RSI chart
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
ax2.plot(df.index, df['RSI'], label='RSI', color='purple')
ax2.axhline(70, color='red', linestyle='--', alpha=0.5)
ax2.axhline(30, color='green', linestyle='--', alpha=0.5)
ax2.set_ylabel("RSI Value")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

# Show last few rows
print(df.tail())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------
# 1. Create or load OHLC data
# ---------------------------------------------------
dates = pd.date_range(start='2022-01-01', periods=200)
high = np.random.normal(0, 1, 200).cumsum() + 110
low = high - np.random.uniform(1, 3, 200)
close = low + np.random.uniform(0, 3, 200)

df = pd.DataFrame({
    'High': high,
    'Low': low,
    'Close': close
}, index=dates)

# ---------------------------------------------------
# 2. Compute Stochastic Oscillator manually (since TA-Lib might not be available)
# ---------------------------------------------------
def compute_stochastic(high, low, close, k_period=14, d_period=3, slowing=3):
    # %K calculation
    lowest_low = low.rolling(window=k_period).min()
    highest_high = high.rolling(window=k_period).max()
    
    # Fast %K
    fastk = 100 * ((close - lowest_low) / (highest_high - lowest_low))
    
    # Slow %K (with slowing)
    slowk = fastk.rolling(window=slowing).mean()
    
    # %D is SMA of slow %K
    slowd = slowk.rolling(window=d_period).mean()
    
    return slowk, slowd

# Apply stochastic calculation
df['slowk'], df['slowd'] = compute_stochastic(
    df['High'], 
    df['Low'], 
    df['Close'], 
    k_period=14, 
    d_period=3, 
    slowing=3
)

# ---------------------------------------------------
# 3. Generate Buy/Sell Signals
# ---------------------------------------------------
df['signal'] = None

for i in range(1, len(df)):
    # SELL: slowk crosses below slowd above 80
    if df['slowk'].iloc[i-1] > df['slowd'].iloc[i-1] and \
       df['slowk'].iloc[i] < df['slowd'].iloc[i] and \
       df['slowk'].iloc[i] > 80:
        df.loc[df.index[i], 'signal'] = 'Sell'

    # BUY: slowk crosses above slowd below 20
    if df['slowk'].iloc[i-1] < df['slowd'].iloc[i-1] and \
       df['slowk'].iloc[i] > df['slowd'].iloc[i] and \
       df['slowk'].iloc[i] < 20:
        df.loc[df.index[i], 'signal'] = 'Buy'

# ---------------------------------------------------
# 4. Plot Price + Stochastic Oscillator
# ---------------------------------------------------
plt.figure(figsize=(14, 8))

# Price chart
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df.index, df['Close'], label='Close Price')
ax1.set_title("Price and Stochastic Oscillator")
ax1.grid(True)
ax1.legend()

# Stochastic chart
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
ax2.plot(df.index, df['slowk'], label='%K (slowk)', color='blue')
ax2.plot(df.index, df['slowd'], label='%D (slowd)', color='orange')
ax2.axhline(80, color='red', linestyle='--', alpha=0.5)
ax2.axhline(20, color='green', linestyle='--', alpha=0.5)
ax2.set_ylabel("Stochastic Value")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 5. Show last rows with signals
# ---------------------------------------------------
print(df[['High', 'Low', 'Close', 'slowk', 'slowd', 'signal']].tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------
# 1. Create or load OHLCV data
# ---------------------------------------------------
dates = pd.date_range(start='2022-01-01', periods=200)
high = np.random.normal(0, 1, 200).cumsum() + 110
low = high - np.random.uniform(1, 3, 200)
close = low + np.random.uniform(0, 3, 200)
volume = np.random.randint(10000, 50000, 200)

df = pd.DataFrame({
    'High': high,
    'Low': low,
    'Close': close,
    'Volume': volume
}, index=dates)

# ---------------------------------------------------
# 2. Money Flow Multiplier (MFM)
# ---------------------------------------------------
df['mfm'] = ((df['Close'] - df['Low']) - (df['High'] - df['Close'])) / (df['High'] - df['Low'])

# Handle division-by-zero cases
df['mfm'] = df['mfm'].replace([np.inf, -np.inf], 0).fillna(0)

# ---------------------------------------------------
# 3. Money Flow Volume (MFV)
# ---------------------------------------------------
df['mfv'] = df['mfm'] * df['Volume']

# ---------------------------------------------------
# 4. Accumulation/Distribution Line (ADL)
# ---------------------------------------------------
df['ADL'] = df['mfv'].cumsum()

# ---------------------------------------------------
# 5. Chaikin Oscillator = EMA(3) - EMA(10)
# ---------------------------------------------------
df['ADL_EMA3'] = df['ADL'].ewm(span=3, adjust=False).mean()
df['ADL_EMA10'] = df['ADL'].ewm(span=10, adjust=False).mean()

df['Chaikin_Osc'] = df['ADL_EMA3'] - df['ADL_EMA10']

# ---------------------------------------------------
# 6. Interpretation: Buying vs Selling Pressure
# ---------------------------------------------------
df['signal'] = np.where(df['Chaikin_Osc'] > 0, 'Buying Pressure', 'Selling Pressure')

# ---------------------------------------------------
# 7. Plot Price + Chaikin Oscillator
# ---------------------------------------------------
plt.figure(figsize=(14, 8))

# Price chart
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df.index, df['Close'], label='Close Price')
ax1.set_title("Price and Chaikin Oscillator")
ax1.grid(True)
ax1.legend()

# Chaikin Oscillator chart
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
ax2.plot(df.index, df['Chaikin_Osc'], label='Chaikin Oscillator', color='purple')
ax2.axhline(0, color='black', linestyle='--', alpha=0.5)
ax2.fill_between(df.index, df['Chaikin_Osc'], 0,
                 where=(df['Chaikin_Osc'] > 0), color='green', alpha=0.3)
ax2.fill_between(df.index, df['Chaikin_Osc'], 0,
                 where=(df['Chaikin_Osc'] < 0), color='red', alpha=0.3)
ax2.set_ylabel("Chaikin Oscillator")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 8. Show last rows
# ---------------------------------------------------
print(df[['High', 'Low', 'Close', 'Volume', 'ADL', 'Chaikin_Osc', 'signal']].tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple


def generate_sample_data(start_date: str = '2022-01-01', 
                         periods: int = 200, 
                         base_price: float = 100) -> pd.DataFrame:
    """
    Generate sample OHLCV data for testing.
    
    Args:
        start_date: Starting date for the time series
        periods: Number of periods (days) to generate
        base_price: Base price to start the random walk from
        
    Returns:
        DataFrame with date index, Close prices and Volume
    """
    dates = pd.date_range(start=start_date, periods=periods)
    close = np.random.normal(0, 1, periods).cumsum() + base_price
    volume = np.random.randint(10000, 50000, periods)
    
    return pd.DataFrame({
        'Close': close,
        'Volume': volume
    }, index=dates)


def calculate_obv(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate On-Balance Volume (OBV) indicator.
    
    OBV adds volume on up days and subtracts volume on down days.
    
    Args:
        df: DataFrame with 'Close' and 'Volume' columns
        
    Returns:
        DataFrame with added 'OBV' column
    """
    # Create a copy to avoid modifying the original dataframe
    result = df.copy()
    result['OBV'] = 0  # initialize
    
    # Get the indices for easier access
    indices = result.index
    
    for i in range(1, len(result)):
        current_idx = indices[i]
        prev_idx = indices[i-1]
        
        if result.loc[current_idx, 'Close'] > result.loc[prev_idx, 'Close']:
            result.loc[current_idx, 'OBV'] = result.loc[prev_idx, 'OBV'] + result.loc[current_idx, 'Volume']
        elif result.loc[current_idx, 'Close'] < result.loc[prev_idx, 'Close']:
            result.loc[current_idx, 'OBV'] = result.loc[prev_idx, 'OBV'] - result.loc[current_idx, 'Volume']
        else:
            result.loc[current_idx, 'OBV'] = result.loc[prev_idx, 'OBV']
            
    return result


def plot_price_and_obv(df: pd.DataFrame) -> Tuple[plt.Figure, plt.Axes]:
    """
    Create a plot showing price and OBV indicator.
    
    Args:
        df: DataFrame with 'Close' and 'OBV' columns
        
    Returns:
        Figure and axes objects for the created plot
    """
    fig = plt.figure(figsize=(14, 8))
    
    # Price chart
    ax1 = plt.subplot(2, 1, 1)
    ax1.plot(df.index, df['Close'], label='Close Price')
    ax1.set_title("Price and On-Balance Volume (OBV)")
    ax1.grid(True)
    ax1.legend()
    
    # OBV chart
    ax2 = plt.subplot(2, 1, 2, sharex=ax1)
    ax2.plot(df.index, df['OBV'], label='OBV', color='purple')
    ax2.set_ylabel("OBV")
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    
    return fig, (ax1, ax2)


# Main execution flow
if __name__ == "__main__":
    # 1. Create sample data
    df = generate_sample_data()
    
    # 2. Calculate OBV
    df = calculate_obv(df)
    
    # 3. Plot the results
    fig, axes = plot_price_and_obv(df)
    plt.show()
    
    # 4. Show last rows
    print(df.tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple


def generate_ohlc_data(start_date: str = '2022-01-01', 
                       periods: int = 200, 
                       base_price: float = 110) -> pd.DataFrame:
    """
    Generate synthetic OHLC price data for testing.
    
    Args:
        start_date: Starting date for the time series
        periods: Number of periods (days) to generate
        base_price: Base price for the high values
        
    Returns:
        DataFrame with date index, High, Low, and Close prices
    """
    dates = pd.date_range(start=start_date, periods=periods)
    high = np.random.normal(0, 1, periods).cumsum() + base_price
    low = high - np.random.uniform(1, 3, periods)
    close = low + np.random.uniform(0, 3, periods)
    
    return pd.DataFrame({
        'High': high,
        'Low': low,
        'Close': close
    }, index=dates)


def calculate_true_range(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate True Range (TR) for the given OHLC data.
    
    True Range is the greatest of:
    1. Current high - current low
    2. |Current high - previous close|
    3. |Current low - previous close|
    
    Args:
        df: DataFrame with 'High', 'Low', and 'Close' columns
        
    Returns:
        DataFrame with added 'TR' column
    """
    result = df.copy()
    result['prev_close'] = result['Close'].shift(1)
    
    # Calculate TR as max(high, prev_close) - min(low, prev_close)
    result['TR'] = result[['High', 'prev_close']].max(axis=1) - result[['Low', 'prev_close']].min(axis=1)
    
    return result


def calculate_atr(df: pd.DataFrame, period: int = 14) -> pd.DataFrame:
    """
    Calculate Average True Range (ATR) using both simple and Wilder's smoothing.
    
    Args:
        df: DataFrame with 'TR' column
        period: Lookback period for ATR calculation
        
    Returns:
        DataFrame with added 'ATR' and 'ATR_wilder' columns
    """
    result = df.copy()
    
    # Simple moving average ATR
    result['ATR'] = result['TR'].rolling(period).mean()
    
    # Wilder's smoothing ATR (more accurate)
    result['ATR_wilder'] = result['TR'].ewm(alpha=1/period, adjust=False).mean()
    
    return result


def plot_price_and_atr(df: pd.DataFrame) -> Tuple[plt.Figure, plt.Axes]:
    """
    Create a plot showing price and ATR indicator.
    
    Args:
        df: DataFrame with 'Close' and 'ATR_wilder' columns
        
    Returns:
        Figure and axes objects for the created plot
    """
    fig = plt.figure(figsize=(14, 8))
    
    # Price chart
    ax1 = plt.subplot(2, 1, 1)
    ax1.plot(df.index, df['Close'], label='Close Price')
    ax1.set_title("Price and Average True Range (ATR)")
    ax1.grid(True)
    ax1.legend()
    
    # ATR chart
    ax2 = plt.subplot(2, 1, 2, sharex=ax1)
    ax2.plot(df.index, df['ATR_wilder'], label='ATR (Wilder)', color='orange')
    ax2.set_ylabel("ATR Value")
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    
    return fig, (ax1, ax2)


def main() -> None:
    """Main function to execute the ATR calculation and visualization."""
    # 1. Generate OHLC data
    df = generate_ohlc_data()
    
    # 2. Calculate True Range
    df = calculate_true_range(df)
    
    # 3. Calculate ATR with both methods
    period = 14
    df = calculate_atr(df, period)
    
    # 4. Plot Price and ATR
    fig, axes = plot_price_and_atr(df)
    plt.show()
    
    # 5. Show last rows
    print(df[['High', 'Low', 'Close', 'TR', 'ATR', 'ATR_wilder']].tail(20))


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, Optional


def generate_ohlc_data(start_date: str = '2025-01-01', 
                       periods: int = 200, 
                       base_price: float = 100) -> pd.DataFrame:
    """
    Generate synthetic OHLC price data for testing.
    
    Args:
        start_date: Starting date for the time series
        periods: Number of periods (days) to generate
        base_price: Base price for the open values
        
    Returns:
        DataFrame with date index, Open, High, Low, and Close prices
    """
    dates = pd.date_range(start=start_date, periods=periods)
    open_ = np.random.normal(0, 1, periods).cumsum() + base_price
    close = open_ + np.random.uniform(-2, 2, periods)
    high = np.maximum(open_, close) + np.random.uniform(0, 0.5, periods)
    low = np.minimum(open_, close) - np.random.uniform(0, 0.5, periods)
    
    return pd.DataFrame({
        'Open': open_,
        'High': high,
        'Low': low,
        'Close': close
    }, index=dates)


def detect_marubozu_candles(df: pd.DataFrame, 
                           shadow_tolerance: float = 0.0) -> pd.DataFrame:
    """
    Detect Marubozu candlestick patterns in OHLC data.
    
    A Marubozu candle has no upper and lower shadows (or very small ones).
    
    Args:
        df: DataFrame with 'Open', 'High', 'Low', and 'Close' columns
        shadow_tolerance: Tolerance for shadow detection (0.0 means exact match)
        
    Returns:
        DataFrame with added columns for Marubozu detection
    """
    result = df.copy()
    
    # Calculate the body high and low
    body_high = result[['Open', 'Close']].max(axis=1)
    body_low = result[['Open', 'Close']].min(axis=1)
    
    # No upper shadow (or within tolerance)
    result['no_upper_shadow'] = (result['High'] - body_high) <= shadow_tolerance
    
    # No lower shadow (or within tolerance)
    result['no_lower_shadow'] = (body_low - result['Low']) <= shadow_tolerance
    
    # Bullish Marubozu (close > open and no shadows)
    result['bullish_marubozu'] = (result['Close'] > result['Open']) & \
                                result['no_upper_shadow'] & result['no_lower_shadow']
    
    # Bearish Marubozu (open > close and no shadows)
    result['bearish_marubozu'] = (result['Open'] > result['Close']) & \
                                result['no_upper_shadow'] & result['no_lower_shadow']
    
    return result


def plot_marubozu_detection(df: pd.DataFrame, 
                           title: str = "Marubozu Candlestick Detection") -> Tuple[plt.Figure, plt.Axes]:
    """
    Create a plot showing price and Marubozu candlestick patterns.
    
    Args:
        df: DataFrame with 'Close', 'bullish_marubozu', and 'bearish_marubozu' columns
        title: Title for the plot
        
    Returns:
        Figure and axes objects for the created plot
    """
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Price line
    ax.plot(df.index, df['Close'], label='Close Price', color='black')
    
    # Bullish Marubozu markers
    bullish_idx = df.index[df['bullish_marubozu']]
    if len(bullish_idx) > 0:
        ax.scatter(bullish_idx,
                  df.loc[bullish_idx, 'Close'],
                  color='blue', marker='^', s=120, label='Bullish Marubozu')
    
    # Bearish Marubozu markers
    bearish_idx = df.index[df['bearish_marubozu']]
    if len(bearish_idx) > 0:
        ax.scatter(bearish_idx,
                  df.loc[bearish_idx, 'Close'],
                  color='red', marker='v', s=120, label='Bearish Marubozu')
    
    ax.set_title(title)
    ax.grid(True)
    ax.legend()
    
    plt.tight_layout()
    
    return fig, ax


def main(shadow_tolerance: float = 0.0) -> None:
    """
    Main function to execute the Marubozu detection and visualization.
    
    Args:
        shadow_tolerance: Tolerance for shadow detection in Marubozu candles
    """
    # 1. Generate OHLC data
    df = generate_ohlc_data()
    
    # 2. Detect Marubozu candles
    df = detect_marubozu_candles(df, shadow_tolerance)
    
    # 3. Plot price with Marubozu markers
    fig, ax = plot_marubozu_detection(df)
    plt.show()
    
    # 4. Show last rows
    columns_to_display = ['Open', 'High', 'Low', 'Close', 
                          'bullish_marubozu', 'bearish_marubozu']
    print(df[columns_to_display].tail(20))


if __name__ == "__main__":
    # You can adjust the shadow tolerance to find more patterns
    # 0.0 means exact match (no shadows at all)
    # 0.1 would allow for small shadows (10% of the typical range)
    main(shadow_tolerance=0.0)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, Optional


def generate_ohlc_data(start_date: str = '2022-01-01', 
                       periods: int = 300, 
                       base_price: float = 100) -> pd.DataFrame:
    """
    Generate synthetic OHLC price data for testing.
    
    Args:
        start_date: Starting date for the time series
        periods: Number of periods (days) to generate
        base_price: Base price for the open values
        
    Returns:
        DataFrame with date index, Open, High, Low, and Close prices
    """
    dates = pd.date_range(start=start_date, periods=periods)
    open_ = np.random.normal(0, 1, periods).cumsum() + base_price
    close = open_ + np.random.uniform(-2, 2, periods)
    high = np.maximum(open_, close) + np.random.uniform(0, 0.3, periods)
    low = np.minimum(open_, close) - np.random.uniform(0, 2, periods)
    
    return pd.DataFrame({
        'Open': open_,
        'High': high,
        'Low': low,
        'Close': close
    }, index=dates)


def calculate_candle_components(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate candlestick components: body size, upper wick, and lower wick.
    
    Args:
        df: DataFrame with 'Open', 'High', 'Low', and 'Close' columns
        
    Returns:
        DataFrame with added 'body', 'upper_wick', and 'lower_wick' columns
    """
    result = df.copy()
    
    # Calculate body size (absolute difference between open and close)
    result['body'] = abs(result['Close'] - result['Open'])
    
    # Calculate upper wick (distance from highest price to top of body)
    result['upper_wick'] = result['High'] - result[['Open', 'Close']].max(axis=1)
    
    # Calculate lower wick (distance from bottom of body to lowest price)
    result['lower_wick'] = result[['Open', 'Close']].min(axis=1) - result['Low']
    
    return result


def detect_hammer_pattern(df: pd.DataFrame, 
                         lower_wick_ratio: float = 2.0,
                         upper_wick_ratio: float = 0.3) -> pd.DataFrame:
    """
    Detect Hammer candlestick patterns in OHLC data.
    
    A Hammer is characterized by:
    - Small body
    - Long lower wick (typically at least 2 body length)
    - Very small or no upper wick
    
    Args:
        df: DataFrame with 'body', 'upper_wick', and 'lower_wick' columns
        lower_wick_ratio: Minimum ratio of lower wick to body
        upper_wick_ratio: Maximum ratio of upper wick to body
        
    Returns:
        DataFrame with added 'hammer' column
    """
    result = df.copy()
    
    # Conditions for Hammer:
    # - Long lower wick (at least lower_wick_ratio  body)
    # - Very small upper wick (at most upper_wick_ratio  body)
    result['hammer'] = (
        (result['lower_wick'] >= lower_wick_ratio * result['body']) &
        (result['upper_wick'] <= result['body'] * upper_wick_ratio)
    )
    
    return result


def plot_hammer_detection(df: pd.DataFrame, 
                         title: str = "Hammer Candlestick Pattern Detection") -> Tuple[plt.Figure, plt.Axes]:
    """
    Create a plot showing price and Hammer pattern markers.
    
    Args:
        df: DataFrame with 'Close' and 'hammer' columns
        title: Title for the plot
        
    Returns:
        Figure and axes objects for the created plot
    """
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Price line
    ax.plot(df.index, df['Close'], label='Close Price', color='black')
    
    # Mark Hammer signals
    hammer_idx = df.index[df['hammer']]
    if len(hammer_idx) > 0:
        ax.scatter(hammer_idx,
                  df.loc[hammer_idx, 'Close'],
                  color='green', marker='^', s=150, label='Hammer (Bullish Reversal)')
    
    ax.set_title(title)
    ax.grid(True)
    ax.legend()
    
    plt.tight_layout()
    
    return fig, ax


def main(lower_wick_ratio: float = 2.0, upper_wick_ratio: float = 0.3) -> None:
    """
    Main function to execute the Hammer pattern detection and visualization.
    
    Args:
        lower_wick_ratio: Minimum ratio of lower wick to body
        upper_wick_ratio: Maximum ratio of upper wick to body
    """
    # 1. Generate OHLC data
    df = generate_ohlc_data()
    
    # 2. Calculate candle components
    df = calculate_candle_components(df)
    
    # 3. Detect Hammer patterns
    df = detect_hammer_pattern(df, lower_wick_ratio, upper_wick_ratio)
    
    # 4. Plot price with Hammer markers
    fig, ax = plot_hammer_detection(df)
    plt.show()
    
    # 5. Show last rows
    columns_to_display = ['Open', 'High', 'Low', 'Close', 
                          'body', 'upper_wick', 'lower_wick', 'hammer']
    print(df[columns_to_display].tail(20))


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from typing import Dict, Tuple, List, Optional, Any


def compute_rsi(series: pd.Series, period: int = 21) -> pd.Series:
    """
    Calculate the Relative Strength Index (RSI) for a price series.
    
    Args:
        series: Price series (typically close prices)
        period: Lookback period for RSI calculation
        
    Returns:
        Series containing RSI values
    """
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi


def prepare_indicators(data: pd.DataFrame, 
                      sma_period: int = 200, 
                      rsi_period: int = 21) -> pd.DataFrame:
    """
    Calculate technical indicators for the trading strategy.
    
    Args:
        data: DataFrame with at least a 'Close' column
        sma_period: Period for Simple Moving Average calculation
        rsi_period: Period for RSI calculation
        
    Returns:
        DataFrame with added indicator columns
    """
    result = data.copy()
    
    # Calculate 200-day SMA
    result['SMA200'] = result['Close'].rolling(sma_period).mean()
    
    # Calculate 21-day RSI
    result['RSI'] = compute_rsi(result['Close'], rsi_period)
    
    return result


def generate_signals(data: pd.DataFrame, 
                    rsi_buy_threshold: float = 70.0,
                    rsi_sell_threshold: float = 50.0) -> pd.DataFrame:
    """
    Generate trading signals based on price, SMA, and RSI.
    
    Strategy:
    - BUY when price > SMA200 and RSI > rsi_buy_threshold
    - SELL when RSI < rsi_sell_threshold
    - Otherwise maintain previous position
    
    Args:
        data: DataFrame with 'Close', 'SMA200', and 'RSI' columns
        rsi_buy_threshold: RSI threshold for buy signals
        rsi_sell_threshold: RSI threshold for sell signals
        
    Returns:
        DataFrame with added 'position' column (1 = long, 0 = cash)
    """
    result = data.copy()
    result['position'] = 0
    
    for i in range(1, len(result)):
        price = result['Close'].iloc[i]
        sma200 = result['SMA200'].iloc[i]
        rsi = result['RSI'].iloc[i]
        prev_pos = result['position'].iloc[i-1]
        
        # BUY condition
        if price > sma200 and rsi > rsi_buy_threshold:
            result.loc[result.index[i], 'position'] = 1
        
        # SELL condition
        elif rsi < rsi_sell_threshold:
            result.loc[result.index[i], 'position'] = 0
        
        # Maintain previous position
        else:
            result.loc[result.index[i], 'position'] = prev_pos
    
    return result


def calculate_performance(data: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate performance metrics for the trading strategy.
    
    Args:
        data: DataFrame with 'Close' and 'position' columns
        
    Returns:
        DataFrame with added performance columns
    """
    result = data.copy()
    
    # Calculate price differences
    result['price_diff'] = result['Close'].diff()
    
    # Calculate strategy P&L
    result['strategy_pnl'] = result['position'].shift(1) * result['price_diff']
    
    # Calculate cumulative P&L
    result['cumulative_pnl'] = result['strategy_pnl'].cumsum()
    
    # Identify trades
    result['trade'] = result['position'].diff().fillna(0)
    
    return result


def backtest_crypto(symbol: str, 
                   start_date: str, 
                   end_date: str,
                   sma_period: int = 200,
                   rsi_period: int = 21,
                   rsi_buy_threshold: float = 70.0,
                   rsi_sell_threshold: float = 50.0) -> Tuple[pd.DataFrame, int]:
    """
    Run a complete backtest for a cryptocurrency using the SMA+RSI strategy.
    
    Args:
        symbol: Cryptocurrency symbol (e.g., "BTC-USD")
        start_date: Start date for backtest
        end_date: End date for backtest
        sma_period: Period for Simple Moving Average
        rsi_period: Period for RSI calculation
        rsi_buy_threshold: RSI threshold for buy signals
        rsi_sell_threshold: RSI threshold for sell signals
        
    Returns:
        Tuple containing:
        - DataFrame with all calculations
        - Number of trades executed
    """
    # Download historical data
    data = yf.download(symbol, start=start_date, end=end_date)
    data = data[['Close']].copy()
    
    # Calculate indicators
    data = prepare_indicators(data, sma_period, rsi_period)
    
    # Generate trading signals
    data = generate_signals(data, rsi_buy_threshold, rsi_sell_threshold)
    
    # Calculate performance
    data = calculate_performance(data)
    
    # Count trades
    trades = data['trade'].abs().sum()
    
    return data, int(trades)


def plot_backtest_results(data: pd.DataFrame, 
                         symbol: str,
                         title: Optional[str] = None) -> Tuple[plt.Figure, plt.Axes]:
    """
    Create a plot showing the cumulative P&L of the trading strategy.
    
    Args:
        data: DataFrame with 'cumulative_pnl' column
        symbol: Cryptocurrency symbol for labeling
        title: Optional title for the plot
        
    Returns:
        Figure and axes objects for the created plot
    """
    fig, ax = plt.subplots(figsize=(14, 8))
    
    ax.plot(data.index, data['cumulative_pnl'], label=f"Strategy P&L ({symbol})")
    
    if title is None:
        title = f"Trend + RSI Strategy Backtest ({symbol})"
    
    ax.set_title(title)
    ax.grid(True)
    ax.legend()
    
    plt.tight_layout()
    
    return fig, ax


def run_multiple_backtests(symbols: List[str], 
                          start_date: str, 
                          end_date: str,
                          **strategy_params) -> Dict[str, Dict[str, Any]]:
    """
    Run backtests for multiple cryptocurrencies.
    
    Args:
        symbols: List of cryptocurrency symbols
        start_date: Start date for backtests
        end_date: End date for backtests
        **strategy_params: Additional parameters for the backtest function
        
    Returns:
        Dictionary with results for each symbol
    """
    results = {}
    
    for coin in symbols:
        try:
            df, trades = backtest_crypto(coin, start_date, end_date, **strategy_params)
            total_return = df['cumulative_pnl'].iloc[-1] if not df.empty else 0
            
            results[coin] = {
                "df": df,
                "trades": trades,
                "total_return": total_return
            }
            
            print(f"{coin}: Trades={trades}, Total Return={total_return:.2f}")
            
        except Exception as e:
            print(f"Error processing {coin}: {str(e
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


class CryptoBacktester:
    def __init__(self, sma_period=200, rsi_period=21, rsi_buy_threshold=70, rsi_sell_threshold=50):
        """
        Initialize the backtester with strategy parameters.
        
        Args:
            sma_period: Period for Simple Moving Average calculation
            rsi_period: Period for Relative Strength Index calculation
            rsi_buy_threshold: RSI threshold for buy signals
            rsi_sell_threshold: RSI threshold for sell signals
        """
        self.sma_period = sma_period
        self.rsi_period = rsi_period
        self.rsi_buy_threshold = rsi_buy_threshold
        self.rsi_sell_threshold = rsi_sell_threshold
    
    @staticmethod
    def compute_rsi(series, period=21):
        """
        Calculate Relative Strength Index for a price series.
        
        Args:
            series: Price series (pandas Series)
            period: RSI calculation period
            
        Returns:
            pandas Series containing RSI values
        """
        delta = series.diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)

        avg_gain = gain.rolling(period).mean()
        avg_loss = loss.rolling(period).mean()

        rs = avg_gain / avg_loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    def calculate_indicators(self, data):
        """
        Calculate technical indicators for the strategy.
        
        Args:
            data: DataFrame with price data
            
        Returns:
            DataFrame with added indicator columns
        """
        df = data.copy()
        df['SMA200'] = df['Close'].rolling(self.sma_period).mean()
        df['RSI21'] = self.compute_rsi(df['Close'], self.rsi_period)
        return df
    
    def generate_signals(self, data):
        """
        Generate trading signals based on indicators.
        
        Args:
            data: DataFrame with price and indicator data
            
        Returns:
            DataFrame with added position column
        """
        df = data.copy()
        df['position'] = 0
        
        # Vectorized operations where possible
        for i in range(1, len(df)):
            price = df['Close'].iloc[i]
            sma = df['SMA200'].iloc[i]
            rsi = df['RSI21'].iloc[i]
            prev_pos = df['position'].iloc[i-1]
            
            # Trading logic
            if price > sma and rsi > self.rsi_buy_threshold:
                df.loc[df.index[i], 'position'] = 1
            elif rsi < self.rsi_sell_threshold:
                df.loc[df.index[i], 'position'] = 0
            else:
                df.loc[df.index[i], 'position'] = prev_pos
                
        return df
    
    def calculate_performance(self, data):
        """
        Calculate strategy performance metrics.
        
        Args:
            data: DataFrame with position data
            
        Returns:
            DataFrame with added performance columns
        """
        df = data.copy()
        df['price_diff'] = df['Close'].diff()
        df['strategy_pnl'] = df['position'].shift(1) * df['price_diff']
        df['cumulative_pnl'] = df['strategy_pnl'].cumsum()
        
        # Count trades
        trades = df['position'].diff().abs().sum()
        
        return df, int(trades)
    
    def backtest(self, symbol, start_date, end_date):
        """
        Run a complete backtest for a cryptocurrency.
        
        Args:
            symbol: Cryptocurrency symbol
            start_date: Start date for backtest
            end_date: End date for backtest
            
        Returns:
            Tuple of (DataFrame with backtest results, number of trades)
        """
        # Download data
        data = yf.download(symbol, start=start_date, end=end_date)
        
        if data.empty:
            return pd.DataFrame(), 0
        
        # Keep only Close prices initially
        data = data[['Close']].copy()
        
        # Apply strategy steps
        data = self.calculate_indicators(data)
        data = self.generate_signals(data)
        data, trades = self.calculate_performance(data)
        
        return data, trades


import ccxt
import pandas as pd
import numpy as np
from datetime import datetime
import time


class CryptoBacktester:
    def __init__(self, exchange_id='binance', sma_period=200, rsi_period=21, 
                 rsi_buy_threshold=70, rsi_sell_threshold=50, timeframe='1d'):
        """
        Initialize the backtester with strategy parameters.
        
        Args:
            exchange_id: CCXT exchange ID (default: 'binance')
            sma_period: Period for Simple Moving Average calculation
            rsi_period: Period for Relative Strength Index calculation
            rsi_buy_threshold: RSI threshold for buy signals
            rsi_sell_threshold: RSI threshold for sell signals
            timeframe: Timeframe for data (e.g., '1d', '4h', '1h')
        """
        self.exchange = getattr(ccxt, exchange_id)()
        self.sma_period = sma_period
        self.rsi_period = rsi_period
        self.rsi_buy_threshold = rsi_buy_threshold
        self.rsi_sell_threshold = rsi_sell_threshold
        self.timeframe = timeframe
    
    @staticmethod
    def compute_rsi(series, period=21):
        """
        Calculate Relative Strength Index for a price series.
        
        Args:
            series: Price series (pandas Series)
            period: RSI calculation period
            
        Returns:
            pandas Series containing RSI values
        """
        delta = series.diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)

        avg_gain = gain.rolling(period).mean()
        avg_loss = loss.rolling(period).mean()

        rs = avg_gain / avg_loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    def fetch_ohlcv_data(self, symbol, start_date, end_date):
        """
        Fetch OHLCV data using CCXT.
        
        Args:
            symbol: Trading pair symbol (e.g., 'BTC/USDT')
            start_date: Start date string in 'YYYY-MM-DD' format
            en
```

```python
import pandas as pd
import numpy as np
import yfinance as yf

# -----------------------------------------
# RSI Calculation
# -----------------------------------------
def compute_RSI(series, period=14):
    """Calculate Relative Strength Index (RSI)"""
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()

    # Avoid division by zero
    avg_loss = avg_loss.replace(0, np.finfo(float).eps)
    
    RS = avg_gain / avg_loss
    RSI = 100 - (100 / (1 + RS))
    return RSI

# -----------------------------------------
# Backtest Function
# -----------------------------------------
def backtest_crypto(symbol, start_date, end_date, ma_period, rsi_period,
                    rsi_buy=70, rsi_sell=50, trade_cost=0.000001):
    """
    Backtest a trading strategy based on Moving Average and RSI
    
    Parameters:
    -----------
    symbol : str
        Cryptocurrency symbol (e.g., 'BTC-USD')
    start_date, end_date : str
        Date range for backtesting in 'YYYY-MM-DD' format
    ma_period : int
        Moving average period
    rsi_period : int
        RSI calculation period
    rsi_buy : int
        RSI threshold for buy signals
    rsi_sell : int
        RSI threshold for sell signals
    trade_cost : float
        Trading cost as a fraction (0.000001 = 0.0001% per side)
        
    Returns:
    --------
    DataFrame, int
        Processed dataframe with strategy results and number of trades
    """
    # Download data
    data = yf.download(symbol, start=start_date, end=end_date)
    if data.empty:
        print(f"No data available for {symbol}")
        return None, 0

    # Create working dataframe
    df = data[['Close']].copy()

    # Calculate indicators
    df['MA'] = df['Close'].rolling(ma_period).mean()
    df['RSI'] = compute_RSI(df['Close'], rsi_period)

    # Initialize columns
    df['position'] = 0
    df['trade_cost'] = 0

    # Trading logic
    for i in range(1, len(df)):
        price = df['Close'].iloc[i]
        ma = df['MA'].iloc[i]
        rsi = df['RSI'].iloc[i]
        prev_pos = df['position'].iloc[i-1]

        # BUY signal
        if price > ma and rsi > rsi_buy:
            if prev_pos == 0:  # Only apply cost when position changes
                df['trade_cost'].iloc[i] = price * trade_cost
            df['position'].iloc[i] = 1

        # SELL signal
        elif rsi < rsi_sell:
            if prev_pos == 1:  # Only apply cost when position changes
                df['trade_cost'].iloc[i] = price * trade_cost
            df['position'].iloc[i] = 0

        # Hold current position
        else:
            df['position'].iloc[i] = prev_pos

    # Calculate PnL
    df['price_diff'] = df['Close'].diff()
    df['strategy_pnl'] = df['position'].shift(1) * df['price_diff']
    df['strategy_pnl'] = df['strategy_pnl'] - df['trade_cost']
    df['cumulative_pnl'] = df['strategy_pnl'].cumsum()

    # Count trades
    trades = df['position'].diff().abs().sum()

    return df, trades

# -----------------------------------------
# Parameter Sweep Function
# -----------------------------------------
def run_parameter_sweep(symbol, start_date, end_date):
    """
    Run backtest with multiple parameter combinations
    
    Parameters:
    -----------
    symbol : str
        Cryptocurrency symbol
    start_date, end_date : str
        Date range for backtesting
        
    Returns:
    --------
    DataFrame
        Results of all parameter combinations
    """
    results = []
    
    print(f"Running parameter sweep for {symbol} from {start_date} to {end_date}")
    print("-" * 60)

    for ma in range(20, 201, 20):  # 20, 40, 60, ..., 200
        for rsi_period in [14, 21]:
            df, trades = backtest_crypto(
                symbol=symbol,
                start_date=start_date,
                end_date=end_date,
                ma_period=ma,
                rsi_period=rsi_period
            )

            if df is None:
                continue

            total_return = df['cumulative_pnl'].iloc[-1]
            
            # Calculate additional metrics
            sharpe = 0
            if df['strategy_pnl'].std() != 0:
                sharpe = (df['strategy_pnl'].mean() / df['strategy_pnl'].std()) * np.sqrt(252)
            
            max_drawdown = (df['cumulative_pnl'].cummax() - df['cumulative_pnl']).max()

            results.append({
                "MA_Period": ma,
                "RSI_Period": rsi_period,
                "Trades": trades,
                "Total_Return": total_return,
                "Sharpe_Ratio": sharpe,
                "Max_Drawdown": max_drawdown
            })

            print(f"MA={ma:3d} | RSI={rsi_period:2d} | Trades={trades:3d} | Return={total_return:.2f} | Sharpe={sharpe:.2f}")

    return pd.DataFrame(results)

# -----------------------------------------
# Main Execution
# -----------------------------------------
if __name__ == "__main__":
    # Run backtest
    df_results = run_parameter_sweep(
        symbol="BTC-USD",
        start_date="2018-01-01",
        end_date="2023-12-31"  # Updated to a valid end date
    )

    print("\nFinal Results:")
    print("-" * 60)
    
    # Sort by total return
    sorted_results = df_results.sort_values(by="Total_Return", ascending=False)
    print(sorted_results)
    
    # Find best parameters
    best_params = sorted_results.iloc[0]
    print("\nBest Parameters:")
    print(f"MA Period: {best_params['MA_Period']}")
    print(f"RSI Period: {best_params['RSI_Period']}")
    print(f"Total Return: {best_params['Total_Return']:.2f}")
    print(f"Number of Trades: {best_params['Trades']}")
    print(f"Sharpe Ratio: {best_params['Sharpe_Ratio']:.2f}")
    print(f"Max Drawdown: {best_params['Max_Drawdown']:.2f}")
```

```python
import ccxt
import pandas as pd
import numpy as np
import time
from datetime import datetime


# ---------------------------------------------------------
# 1. RSI Calculation
# ---------------------------------------------------------
def compute_RSI(series, period=14):
    """Calculate Relative Strength Index for a price series"""
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()

    RS = avg_gain / avg_loss
    RSI = 100 - (100 / (1 + RS))
    return RSI


# ---------------------------------------------------------
# 2. Fetch OHLCV using CCXT
# ---------------------------------------------------------
def fetch_ohlcv_ccxt(exchange, symbol, timeframe, start_date, end_date):
    """Fetch historical OHLCV data from exchange"""
    # Convert dates to timestamps
    start_ts = int(datetime.strptime(start_date, "%Y-%m-%d").timestamp() * 1000)
    end_ts = int(datetime.strptime(end_date, "%Y-%m-%d").timestamp() * 1000)

    all_data = []
    since = start_ts

    print(f"Fetching {symbol} from {start_date} to {end_date}...")

    # Fetch data in batches
    while since < end_ts:
        try:
            candles = exchange.fetch_ohlcv(symbol, timeframe, since, limit=1000)
            if not candles:
                break

            all_data.extend(candles)
            since = candles[-1][0] + 1
            # Respect exchange rate limits
            time.sleep(exchange.rateLimit / 1000)

        except Exception as e:
            print(f"Error fetching data: {e}")
            break

    # Create DataFrame from fetched data
    df = pd.DataFrame(
        all_data,
        columns=["timestamp", "open", "high", "low", "close", "volume"]
    )

    # Format timestamp and set as index
    df["timestamp"] = pd.to_datetime(df["timestamp"].astype("int64"), unit="ms")
    df.set_index("timestamp", inplace=True)

    return df


# ---------------------------------------------------------
# 3. Backtest Strategy (MA + RSI + Trade Cost)
# ---------------------------------------------------------
def backtest_crypto(df, ma_period, rsi_period, rsi_buy=70, rsi_sell=50, trade_cost=0.000001):
    """
    Backtest a trading strategy using Moving Average and RSI
    
    Parameters:
    - df: DataFrame with OHLCV data
    - ma_period: Moving Average period
    - rsi_period: RSI calculation period
    - rsi_buy: RSI threshold for buy signals
    - rsi_sell: RSI threshold for sell signals
    - trade_cost: Trading fee as a percentage (0.000001 = 0.0001%)
    
    Returns:
    - DataFrame with strategy results
    - Number of trades executed
    """
    df = df.copy()
    
    # Calculate indicators
    df.loc[:, "MA"] = df["close"].rolling(ma_period).mean()
    df.loc[:, "RSI"] = compute_RSI(df["close"], rsi_period)

    # Initialize position and cost columns with correct dtypes
    df.loc[:, "position"] = 0.0  # Initialize as float
    df.loc[:, "trade_cost"] = 0.0  # Initialize as float

    # Implement trading logic
    for i in range(1, len(df)):
        price = df["close"].iloc[i]
        ma = df["MA"].iloc[i]
        rsi = df["RSI"].iloc[i]
        prev_pos = df["position"].iloc[i - 1]

        # BUY condition: price above MA and RSI above threshold
        if price > ma and rsi > rsi_buy:
            if prev_pos == 0:  # Only apply cost when position changes
                df.loc[df.index[i], "trade_cost"] = price * trade_cost
            df.loc[df.index[i], "position"] = 1.0

        # SELL condition: RSI below threshold
        elif rsi < rsi_sell:
            if prev_pos == 1:  # Only apply cost when position changes
                df.loc[df.index[i], "trade_cost"] = price * trade_cost
            df.loc[df.index[i], "position"] = 0.0

        # Hold current position
        else:
            df.loc[df.index[i], "position"] = prev_pos

    # Calculate PnL
    df.loc[:, "price_diff"] = df["close"].diff()
    df.loc[:, "strategy_pnl"] = df["position"].shift(1) * df["price_diff"]
    df.loc[:, "strategy_pnl"] -= df["trade_cost"]  # Subtract trading costs
    df.loc[:, "cumulative_pnl"] = df["strategy_pnl"].cumsum()

    # Count trades
    trades = df["position"].diff().abs().sum()

    return df, int(trades)


# ---------------------------------------------------------
# 4. Parameter Sweep (MA 20200, RSI 14 & 21)
# ---------------------------------------------------------
def run_parameter_sweep(symbol, start_date, end_date, exchange_id="kraken", timeframe="1d"):
    """
    Run parameter sweep to find optimal MA and RSI settings
    
    Parameters:
    - symbol: Trading pair (e.g., "BTC/USD")
    - start_date: Backtest start date (YYYY-MM-DD)
    - end_date: Backtest end date (YYYY-MM-DD)
    - exchange_id: Exchange to use (default: "kraken")
    - timeframe: Candle timeframe (default: "1d")
    
    Returns:
    - DataFrame with results for each parameter combination
    """
    # Initialize exchange
    exchange = getattr(ccxt, exchange_id)()

    # Fetch historical data
    df = fetch_ohlcv_ccxt(exchange, symbol, timeframe, start_date, end_date)
    if df.empty:
        print("No data fetched.")
        return None

    results = []

    # Test different parameter combinations
    for ma in range(20, 201, 20):
        for rsi_period in [14, 21]:
            # Run backtest with current parameters
            df_bt, trades = backtest_crypto(
                df=df,
                ma_period=ma,
                rsi_period=rsi_period
            )

            total_return = df_bt["cumulative_pnl"].iloc[-1]

            # Store results
            results.append({
                "MA": ma,
                "RSI_Period": rsi_period,
                "Trades": trades,
                "Total_Return": total_return
            })

            # Print progress
            print(f"{symbol} | MA={ma} | RSI={rsi_period} | Trades={trades} | Return={total_return:.2f}")

    return pd.DataFrame(results)


# ---------------------------------------------------------
# 5. Run Example
# ---------------------------------------------------------
if __name__ == "__main__":
    results = run_parameter_sweep(
        symbol="BTC/USD",  # Kraken typically uses USD instead of USDT
        start_date="2018-01-01",
        end_date="2024-12-31",
        exchange_id="kraken",
        timeframe="1d"
    )

    print("\nFinal Results:")
    print(results)
    
    # Find best parameter combination
    if results is not None and not results.empty:
        best_result = results.loc[results["Total_Return"].idxmax()]
        print("\nBest Parameters:")
        print(f"MA Period: {best_result['MA']}")
        print(f"RSI Period: {best_result['RSI_Period']}")
        print(f"Total Return: {best_result['Total_Return']:.2f}")
        print(f"number of Trades: {best_result['Trades']}")
            
        
```

```python
import pandas as pd
import requests
import time

def get_kraken_ohlc(pair="BTCUSD", interval=60, since=None):
    """
    Fetch OHLC data from Kraken API
    
    Parameters:
    pair (str): Trading pair
    interval (int): Time frame interval in minutes (1, 5, 15, 30, 60, 240, 1440, 10080, 21600)
    since (int): Return committed OHLC data since given timestamp
    
    Returns:
    pandas.DataFrame: OHLC data
    """
    # Kraken API endpoint for OHLC data
    url = "https://api.kraken.com/0/public/OHLC"
    
    # Parameters for the API request
    params = {
        "pair": pair,
        "interval": interval
    }
    
    if since:
        params["since"] = since
    
    # Make the API request
    response = requests.get(url, params=params)
    data = response.json()
    
    # Check for errors
    if len(data['error']) > 0:
        print(f"Error: {data['error']}")
        return None
    
    # Extract the result data for the pair
    result_key = list(data['result'].keys())[0]
    ohlc_data = data['result'][result_key]
    
    # Create DataFrame
    df = pd.DataFrame(ohlc_data, columns=[
        "time", "open", "high", "low", "close", "vwap", "volume", "count"
    ])
    
    # Convert types
    df["time"] = df["time"].astype(float)
    for col in ["open", "high", "low", "close", "vwap", "volume"]:
        df[col] = df[col].astype(float)
    df["count"] = df["count"].astype(int)
    
    # Convert UNIX timestamp to datetime
    df["timestamp"] = pd.to_datetime(df["time"], unit="s")
    
    # Sort by timestamp
    df = df.sort_values("timestamp")
    
    # Set timestamp as index
    df = df.set_index("timestamp")
    
    return df

# Example usage: Get hourly BTC/USD data for the last day
# Note: Kraken uses XBT instead of BTC in their API
df = get_kraken_ohlc(pair="XBTUSD", interval=60)

# Display the first few rows
print(df.head())

# If you want to save this data to a CSV for future use
df.to_csv("BTCUSD_60.csv")
```
